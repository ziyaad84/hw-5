```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_table = pd.merge(mouse_df, clinical_df, on="Mouse ID")

# Display the data table for preview
merge_table.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
merge_table_1 = merge_table[['Drug','Timepoint', 'Tumor Volume (mm3)']]
grouped_mouse = merge_table_1.groupby(['Drug','Timepoint']).mean()

# Convert to DataFrame
group_mouse_df = grouped_mouse.reset_index('Timepoint')

# Preview DataFrame
group_mouse_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sems_1 =  merge_table_1.groupby(['Drug','Timepoint']).sem()

# Convert to DataFrame
sems_df = sems_1.reset_index('Timepoint')

# Preview DataFrame
sems_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_reformat_df = group_mouse_df.pivot_table('Tumor Volume (mm3)' ,'Timepoint', 'Drug')
sems_reformat_df = sems_df.pivot_table('Tumor Volume (mm3)','Timepoint' , 'Drug')

# Preview that Reformatting worked
mouse_reformat_df.head()
sems_reformat_df.head()
```

```python
# Generate the Plot (with Error Bars)
fig_1 = mouse_reformat_df.plot(kind='line')
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()

fig_2 = sems_reformat_df.plot(kind='line')    
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3) Sems")
plt.grid()

# Save the Figure

fig = fig_1.get_figure()
fig.savefig("figure.jpeg")


fig_sems = fig_2.get_figure()
fig_sems.savefig("figure_2.jpeg")

# Show the Figure
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame

merge_table_2 = merge_table[['Drug','Timepoint', 'Metastatic Sites']]
grouped_meta = merge_table_2.groupby(['Drug','Timepoint']).mean()

# Convert to DataFrame
group_meta_df = grouped_meta.reset_index('Timepoint')

# Preview DataFrame
group_meta_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame
met_1 =  merge_table_2.groupby(['Drug','Timepoint']).std()

# Convert to DataFrame
met_df = met_1.reset_index('Timepoint')

# Preview DataFrame
met_df.head()


```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked

met_reformat_df = group_meta_df.pivot_table('Metastatic Sites' ,'Timepoint', 'Drug')
met_std_df = met_df.pivot_table('Metastatic Sites','Timepoint' , 'Drug')

# Preview that Reformatting worked
met_reformat_df.head()
met_std_df.head()

```

```python
# Generate the Plot (with Error Bars)

# Save the Figure

# Show the Figure

fig_3 = met_reformat_df.plot(kind='line')
plt.title("Metastatic Spread Through Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met Sites")
plt.grid()

fig_4 = sems_reformat_df.plot(kind='line')    
plt.title("Metastatic Spread Through Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Meta Stdev")
plt.grid()

# Save the Figure

fig5 = fig_3.get_figure()
fig5.savefig("figure.jpeg")


fig_std = fig_4.get_figure()
fig_std.savefig("figure_2.jpeg")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
#merge_table_count = merge_table[['Drug','Timepoint']].count()["Mouse ID"]

#merge_table_count = pd.DataFrame({"Mouse Count": merge_table_count})

# Preview DataFrame
#merge_table_count.head()
mice_count = merge_table.groupby(['Drug', 'Timepoint']).count()['Mouse ID']
mice_count = mice_count.unstack(level = 0)
mice_count

# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)

# Save the Figure

xvals = mice_count.index #timepoints

count = 0 

for c in mice_count:
    yvals = mice_count[c]/mice_count.loc[0,c] * 100 #calculates survival rate for each timepoint
    plt.plot(xvals, 
             yvals, 
             linestyle = '--', 
             )
    count += 1

    lg = plt.legend()
plt.title("Survival Rate During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate %")
plt.grid()

x = plt.show()

x.savefig("count.jpeg")

```

## Summary Bar Graph

```python
tumor_change = (group_mouse_df.loc[45, :] - group_mouse_df.loc[0, :])/group_mouse_df.loc[0, :] * 100
tumor_change
```

```python
#plot options
plt.title('Tumor Volume Change over 45 Day Treatment', fontdict = {'fontsize': 12, 'fontweight': 'bold'})
plt.ylabel('Survival Rate (%)')
plt.axhline(y=0, color = 'black') #adds a horizontal line at zero
xlabels = tumor_change.index
#puts drug names as labels
plt.xticks(np.arange(len(xlabels)), xlabels)

#graphs % change of tumor as bar graph
plt.bar(np.arange(4), 
        tumor_change, 
        # colors each bar appropriately based on + or - values by using a conditional
        color = ['red' if tumor_change[r] > 0 else 'green' for r in np.arange(len(xlabels))]
       )

#count for x coordinate location of value label
count = 0

#loops through tumor change data and places the value in the correct position based on + or - values
for r in tumor_change:
    if r < 0:
        y_coor = -3.5
    else:
        y_coor = 2
    plt.text(count, y_coor, str(round(r, 1)) + '%', ha = 'center', color = 'white')
    count += 1
    
plt.show()
```
